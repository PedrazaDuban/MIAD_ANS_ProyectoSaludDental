# Población con Acceso a Servicio de Salud Dental en la Clínica Universidad Autónoma de Sinaloa


## Librerias


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1.Carga de datos

En la carpeta `Data` se encuentra el archivo `ServicioSocialDental2.xls` dentro del repositorio. Acontinuación se realiza el cargue mediante la función `read_excel`.

```python

df = pd.read_excel("../Data/ServicioSocialDental2.xlsx")
df.head()
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
#Histogramas
df.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
# Correlogramas
sns.pairplot(df, kind="scatter")
plt.show()
```

```python
# plotting correlation heatmap
dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
  
# displaying heatmap
plt.show()
```
